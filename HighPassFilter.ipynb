```python
#import library
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Open File + Declare variable
img0 = cv2.imread('/content/Input-test-images-a-Cameraman-grayscale-b-grayscale-Lena-and-c-color-Lena-All_Q640.jpg')
gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)
img = cv2.GaussianBlur(gray,(3,3), 0)

laplacian = cv2.Laplacian(img, cv2.CV_64F)
sobelx = cv2.Sobel(img, cv2.CV_64F, 1,0, ksize = 5)
sobelx = cv2.Sobel(img, cv2.CV_64F, 0,1, ksize = 5)
```

```python
#ploting

#original image
plt.subplot(2,2,1), plt.imshow(img, cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])

#laplacian image
plt.subplot(2,2,2), plt.imshow(img, cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])

#Sobel X image
plt.subplot(2,2,3), plt.imshow(img, cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])

#Sobel Y image
plt.subplot(2,2,4), plt.imshow(img, cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```
